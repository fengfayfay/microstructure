```python
from math import cos, acos, radians, degrees, sqrt
from vec3 import Vec3
from ray import Ray
from vgroove import VGroove
from mc import uniform
from stats import histogram
import numpy as np

np.set_printoptions(formatter={'float': '{: 0.3f}'.format})

N = 10000

def mc_brdf(cosm, cosi):
    sini = 1-sqrt(cosi*cosi)
    v = VGroove(cosm)
    hits = []
    for i in range(N):
        r = Ray( Vec3(uniform(-1.0, 1.0), 0.0, 0.0), Vec3(sini, 0.0, -cosi) )
        hits += v.trace(r)
    return hits

def brdf(m,i):
    cosm = cos(radians(m))
    cosi = cos(radians(i))
    sini = 1-sqrt(cosi*cosi)
    hits = mc_brdf(cosm, cosi)

    hist = histogram(hits)
    nbins = len(hist)
    print('m={}, cosm={:5.3f}, i={}, cosi={:5.3f}'.format(m,cosm,i,cosi))
    for i in range(nbins):
          print('cosr={:4.2f}'.format(i/nbins),hist[i,:])

#print('m,i')
#for m in [15, 30, 45, 60, 75]:
#    for i in [10,20,30,40,50,60,70,80]:
#        brdf(m,i)

brdf(40,30)
```
