```python
import math
import zipin
import zipinPaper
def compHit(a, b):
    return math.fabs(a[0] + b[0])< .0000001 and math.fabs(a[1]-b[1]) < .0000001 and a[2] == b[2] and a[3] == b[3]
def compHits(a, b):
    if len(a) != len(b):
        return False
    for i in range(len(a)):
        if compHit(a[i], b[i]) == False:
            return False
    return True
for theta in range(10, 89):
    for phi in range(2, 88):
        z1 = zipin.zipin(math.radians(theta), math.radians(phi))
        z2 = zipinPaper.zipin(theta, phi)
        if compHits(z1, z2) == False:
            print(theta, phi)
            print(z1)
            print(z2)
```
